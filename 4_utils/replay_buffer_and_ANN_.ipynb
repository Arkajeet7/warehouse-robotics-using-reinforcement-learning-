```python
import gymnasium as gym
import numpy as np
from gymnasium import spaces
import matplotlib.pyplot as plt
from collections import deque
import random
import torch
from torch import nn
import torch.nn.functional as F
import importlib.util
import pandas as pd
```

```python
class DQN(nn.Module):
    def __init__(self, rows, cols, h1_nodes, out_actions):
        super().__init__()
        # Convolutional layers to process the 2D grid
        input_size = rows * cols  # Flattened input grid

        self.fc1 = nn.Linear(input_size, h1_nodes)
        self.fc2 = nn.Linear(h1_nodes, h1_nodes // 2)
        self.out = nn.Linear(h1_nodes // 2, out_actions)

    def forward(self, x):
        # Flatten input if it's 2D or 3D
        if x.dim() > 2:
            x = x.view(x.size(0), -1)  # (batch, features)
        else:
            x = x.view(1, -1)

        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.out(x)
        return x

# Define memory for Experience Replay
class ReplayMemory():
    def __init__(self, maxlen):
        self.memory = deque([], maxlen=maxlen)

    def append(self, transition):
        self.memory.append(transition)

    def sample(self, sample_size):
        return random.sample(self.memory, sample_size)

    def __len__(self):
        return len(self.memory)
```
